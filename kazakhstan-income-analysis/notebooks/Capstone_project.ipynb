#Расчет показателей

```python
#Динамика доходов

df_1['nominal_growth_%'] = df_1.groupby('Регион')['Nom_Income'].pct_change() * 100
df_1['real_growth_%'] = df_1.groupby('Регион')['Real_Income'].pct_change() * 100

#Сравнение номинальных и реальных насколько инфляция сьедает доходы то же для декомпозиции доходов
df_1['inflation_contribution_%'] = -((df_1['nominal_growth_%'] - df_1['real_growth_%'])).astype(float)

#чтобы показать декомпозицию роста дохода
df_1['nominal_contribution_%'] = df_1['nominal_growth_%']


#Какую долю роста сьела инфляция
df_1['inflation_share_%'] = (
    df_1['inflation_contribution_%'] / df_1['nominal_contribution_%']
) * 100

#Эластичность дохода к инфляции
df_1['elasticity'] = df_1['real_growth_%'] / df_1['CPA']

df_1.fillna(0, inplace=True)

#Доля потребления в доходах
df_1['consumption_share_real'] = df_1['Cons_expenses'] / df_1['Real_Income'] * 100

df_1
```

```python
#для понимания влияния кризисных ситуаций
# Список базовых годов
base_years = [2008, 2015, 2020]

for year in base_years:
    base_dict = (
        df_1[df_1['Год'] == year]
        .set_index('Регион')['Real_Income']
        .to_dict()
    )

    df_1[f'income_index_{year}'] = df_1['Real_Income'] / df_1['Регион'].map(base_dict) * 100

df_1
```

```python
#Баланс: тратят больше или меньше, чем зарабатывают
df_2['balance'] = df_2['Money_Income'] - df_2['Money_Expenses']

#Норма сбережений
df_2['saving_rate'] = (df_2['Money_Income'] - df_2['Money_Expenses']) / df_2['Money_Income'] * 100

#Структура расходов
df_2['consumption_share_total'] = df_2['Cons_expenses'] / df_2['Money_Expenses'] * 100

df_2
```
